# Combine & plot total_loss from multiple subfolders under save_loss

```python
# %% Combine & plot total_loss from multiple subfolders under save_loss
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# ======== USER SETTINGS ========
PARENT_DIR   = "save_loss"          # parent folder that contains the 6 subfolders
SUBFOLDERS   = ["PENUNET_MR","BELLOW_MR","BENDING_JOINT_MR","PENUNET_YEOH","BELLOW_YEOH","BENDING_JOINT_YEOH"]                   # [] -> auto-discover; or provide e.g. ["run1","run2",...]
USE_LOG_Y    = True

FONT_FAMILY  = "Times New Roman"
TITLE_FS = 18; LABEL_FS = 16; TICK_FS = 13; LEGEND_FS = 13

SAVE_BASENAME = "Compare_total_losses_reload"   # files saved to <PARENT_DIR>/<SAVE_BASENAME>.pdf/.png
PNG_DPI = 1080
# =================================

# Fonts & embedding (avoid Type-3 in PDF/PS)
mpl.rcParams.update({
    'font.family':  FONT_FAMILY,
    'text.usetex':  False,
    'pdf.fonttype': 42,
    'ps.fonttype':  42,
    'svg.fonttype': 'none',
})

# Discover subfolders if not specified
# If you want to plot specific six folders in a particular order, set SUBFOLDERS = ["delete_BENDING_YEOH_testt", "run2", "run3", "run4", "run5", "run6"] (order defines which ones get solid vs dashed).
if not SUBFOLDERS:
    SUBFOLDERS = []
    for name in sorted(os.listdir(PARENT_DIR)):
        path = os.path.join(PARENT_DIR, name)
        if os.path.isdir(path) and os.path.exists(os.path.join(path, "total_loss.npy")):
            SUBFOLDERS.append(name)

if not SUBFOLDERS:
    raise RuntimeError(f"No subfolders with total_loss.npy found under: {os.path.abspath(PARENT_DIR)}")

# Style cycle: first 3 solid (red, blue, green), next 3 dashed (red, blue, green)
styles = [
    ("red",   "-"),
    ("blue",  "-"),
    ("green", "-"),
    ("red",   "--"),
    ("blue",  "--"),
    ("green", "--"),
]

# Load and plot
fig, ax = plt.subplots(figsize=(10, 6), dpi=1080)

max_len = 0
loaded = []

for i, sub in enumerate(SUBFOLDERS):
    sub_path = os.path.join(PARENT_DIR, sub)
    tot_path = os.path.join(sub_path, "Total_loss.npy")
    if not os.path.exists(tot_path):
        print(f"Skipping {sub}: missing total_loss.npy")
        continue

    tot = np.load(tot_path)
    iters = np.arange(len(tot))
    color, ls = styles[i % len(styles)]
    label = sub  # legend will show subfolder name

    ax.plot(iters, tot, color=color, linestyle=ls, linewidth=2, label=label)
    max_len = max(max_len, len(tot))
    loaded.append(sub)

if not loaded:
    raise RuntimeError("No curves were loaded (all missing total_loss.npy?).")

# Scales and labels
if USE_LOG_Y:
    ax.set_yscale('log')

ax.set_xlabel('Iteration', fontsize=LABEL_FS, fontname=FONT_FAMILY)
ax.set_ylabel('Loss (log scale)' if USE_LOG_Y else 'Loss',
              fontsize=LABEL_FS, fontname=FONT_FAMILY)
# ax.set_title('Total Loss Across Runs', fontsize=TITLE_FS, fontname=FONT_FAMILY)

# Ensure tick labels use your font
fig.canvas.draw()
ax.tick_params(axis='both', which='both', labelsize=TICK_FS)
for tick in ax.get_xticklabels() + ax.get_yticklabels():
    tick.set_fontname(FONT_FAMILY)

# X-axis from 0 to last-max iteration + small right pad
right = max(0, max_len - 1)
pad = max(1, int(np.ceil(0.03 * max(1, right))))  # ~3% pad; tweak as needed
ax.set_xlim(0, right + pad)

# Legend
ax.legend(loc='upper right', frameon=False,
          prop={'family': FONT_FAMILY, 'size': LEGEND_FS})

fig.tight_layout()

# Save combined figure into the parent folder
os.makedirs(PARENT_DIR, exist_ok=True)
pdf_path = os.path.abspath(os.path.join(PARENT_DIR, SAVE_BASENAME + ".pdf"))
png_path = os.path.abspath(os.path.join(PARENT_DIR, SAVE_BASENAME + ".png"))
plt.savefig(pdf_path, bbox_inches='tight')
plt.savefig(png_path, dpi=PNG_DPI, bbox_inches='tight')
print("Saved to:\n  PDF:", pdf_path, "\n  PNG:", png_path)

plt.show()
```
